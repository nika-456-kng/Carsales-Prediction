```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

print("Shape of scaled X_train:", X_train_scaled.shape)
print("Shape of scaled X_test:", X_test_scaled.shape)
```

```python
scaled_linear_model = LinearRegression()
scaled_linear_model.fit(X_train_scaled, y_train)
```

```python
scaled_linear_predictions = scaled_linear_model.predict(X_test_scaled)
```

```python
print("Linear Regression Model Evaluation (trained and evaluated on scaled data):")
modelresult(scaled_linear_predictions)
```

```python
from sklearn.svm import SVR
svr_model = SVR()
svr_model.fit(X_train_scaled, y_train)
```

```python
svr_predictions = svr_model.predict(X_test_scaled)
```

```python
print("Support Vector Regressor Model Evaluation:")
modelresult(svr_predictions)
```

```python
knn_model = KNeighborsRegressor()
knn_model.fit(X_train_scaled, y_train)
```

```python
knn_predictions = knn_model.predict(X_test_scaled)

print("\nKNeighbors Regressor Model Evaluation:")
modelresult(knn_predictions)
```

```python
model_performance = pd.DataFrame({
    'Model': ['Linear Regression', 'Random Forest Regressor', 'SVR', 'KNN Regressor'],
    'Mean Absolute Error': [
        mean_absolute_error(y_test, scaled_linear_predictions),
        mean_absolute_error(y_test, rf_predictions),
        mean_absolute_error(y_test, svr_predictions),
        mean_absolute_error(y_test, knn_predictions)
    ],
    'Root Mean Squared Error': [
        np.sqrt(mean_squared_error(y_test, scaled_linear_predictions)),
        np.sqrt(mean_squared_error(y_test, rf_predictions)),
        np.sqrt(mean_squared_error(y_test, svr_predictions)),
        np.sqrt(mean_squared_error(y_test, knn_predictions))
    ]
})

display(model_performance)
```

```python

model_performance_melted = model_performance.melt(
    id_vars='Model',
    value_vars=['Mean Absolute Error', 'Root Mean Squared Error'],
    var_name='Metric',
    value_name='Error Value'
)

plt.figure(figsize=(12, 6))
sns.barplot(
    data=model_performance_melted,
    x='Model',
    y='Error Value',
    hue='Metric',
    palette='viridis'
)
plt.title('Comparison of Model Performance (MAE and RMSE)')
plt.xlabel('Model')
plt.ylabel('Error Value')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
